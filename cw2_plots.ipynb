### a). Exponential Decay ###

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rewards_exp1 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP1/mean_rewards_DQN_exp1.txt")
rewards_exp2 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP2/mean_rewards_DQN_exp2.txt")
rewards_exp3 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP3/mean_rewards_DQN_exp3.txt")
rewards_exp4 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP4/mean_rewards_DQN_exp4.txt")

std_exp1 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP1/std_rewards_DQN_exp1.txt")
std_exp2 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP2/std_rewards_DQN_exp2.txt")
std_exp3 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP3/std_rewards_DQN_exp3.txt")
std_exp4 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP4/std_rewards_DQN_exp4.txt")

decay_curve1 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP1/exp_decay1.txt")
decay_curve2 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP2/exp_decay2.txt")
decay_curve3 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP3/exp_decay3.txt")
decay_curve4 = np.loadtxt("Q2/Exponential Epsilon Decay/EXP4/exp_decay4.txt")

stds_exp = [std_exp1, std_exp2, std_exp3, std_exp4]
mean_rewards_exp = [rewards_exp1, rewards_exp2, rewards_exp3, rewards_exp4]
decay_curves = [decay_curve1, decay_curve2, decay_curve3, decay_curve4]
episodes = np.arange(len(rewards_exp1))
colors = ["red", "orange", "green", "blue"]

fig, ax = plt.subplots(2, 1, figsize=(10,14))
ax[0].grid()
ax[1].grid()
for i, (mean, std, decay_curve) in enumerate(zip(mean_rewards_exp, stds_exp, decay_curves)):
    ax[0].plot(episodes, mean, color=colors[i], label=f"Method #{i+1}")
    ax[0].fill_between(episodes, mean-std, mean+std, color=colors[i], alpha=0.4)
    ax[1].plot(episodes, decay_curve, color=colors[i], label=f"Method #{i+1}")
    plt.subplots_adjust(wspace=0, hspace=0)


ax[0].set_ylabel("Mean Total Non-Discounted Reward")
ax[0].set_ylim((0,550))
ax[0].axhline(y=500, color='black', linestyle='-', label="Maximum reward")

ax[1].set_xlabel("Episode")
ax[1].set_ylabel("Epsilon")

ax[0].legend(loc="upper left");
```

```python
def find_perc_episode(mean, absolute_val_array, episodes_array):
    finished=False
    while not finished:
        smallest_difference_index = absolute_val_array.argmin()

        if mean[smallest_difference_index] < np.max(mean[:smallest_difference_index]):
            absolute_val_array = np.delete(absolute_val_array, smallest_difference_index)
        else:
            break

    return episodes_array[smallest_difference_index]


final_performance_value = np.mean(mean_rewards_rbd[-100:])
absolute_val_array = np.abs(mean_rewards_rbd - 0.9*final_performance_value)
smallest_difference_index = absolute_val_array.argmin()

episode_90_perc = find_perc_episode(mean_rewards_rbd, absolute_val_array, episodes)
print("Final performance value = ", final_performance_value)
print("Episode at which 90 percent of final performance is achieved: ", episode_90_perc)
```

## Q2.2 Effect of Replay Buffer on Variablity of Total Reward ## 

```python
buffers_array = np.array([1000, 10000, 100000, 500000, 1000000])
mean_buffers = []
stds_buffers = []

for buff in buffers_array:
    mean_buffer = np.loadtxt(f"Q2/Replay Buffer/mean_rewards_DQN_replay_{buff}.txt")
    stds_buffer = np.loadtxt(f"Q2/Replay Buffer/std_rewards_DQN_replay_{buff}.txt")
    mean_buffers.append(mean_buffer)
    stds_buffers.append(stds_buffer)

episodes = np.arange(len(mean_buffers[0]))
```

```python
stds_67 = []

for mean, stds in zip(mean_buffers, stds_buffers):
    final_performance_value = np.max(mean)
    absolute_val_array = np.abs(mean - 0.67*final_performance_value)
    smallest_difference_index = absolute_val_array.argmin()

    try:
        episode = find_perc_episode(mean, absolute_val_array, episodes)
    except ValueError:
        episode = episodes[0]

    std = np.average(stds[episode:])
    stds_67.append(std)
```

```python

plt.figure(figsize=(10,7))
plt.grid()
plt.plot(buffers_array, stds_67, "--ro")
plt.xlabel("Replay Buffer Size", fontsize=13)
plt.ylabel("Standard Deviation from 67$\%$ of Maximum Achieved Reward", fontsize=13)
plt.xscale("log")
#plt.axvline(x=500, color="black", linestyle="--", label="Maximum episode duration")
#plt.legend()
plt.show();
```
